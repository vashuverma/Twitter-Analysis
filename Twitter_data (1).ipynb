```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("users.csv",encoding= 'unicode_escape')
```

```python
df2 = pd.read_csv("tweets.csv",encoding= 'unicode_escape')
```

```python
df1.head()
```

```python
df2.head()
```

```python
dataset = pd.concat([df1, df2], axis=1)
```

```python
dataset.head()
```

```python
columns_to_keep = ["id", "name", "followers_count", "lang", "time_zone", "text","retweet_count",]
dataset = dataset[columns_to_keep]
```

```python
dataset.head()
```

REMOVE NULL TIME ZONES

```python
language_to_timezone = {
    'ja': 'Tokyo',
    'en': 'Easter Time (US & Canada)',
    'fr': 'France'
}
mask = dataset['time_zone'].isna()
dataset.loc[mask, 'time_zone'] = dataset.loc[mask, 'lang'].map(language_to_timezone)
```

```python
dataset.head()
```

```python
file_path = 'dataset.csv'
dataset.to_csv(file_path, index=False)
```
